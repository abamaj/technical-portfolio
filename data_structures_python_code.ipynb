```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
import csv
```

```python
a = []
b = []
x = []
plt.title("Linked List Average Insert Time and Average Search Time (DataSet A)")
plt.xlabel("Iteration")
plt.ylabel("Time (milliseconds)")
with open('DataSetALinkedLIstInsert.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter = ',')
    for records in readCSV:
        a.append(float(records[0]))
with open('DataSetALinkedLIstsearch.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter = ',')
    for records in readCSV:
        b.append(float(records[0]))
    for numbers in range(0, 400):
        x.append(numbers)
plt.xlim(0, 400)
plt.ylim(0, .0002)
plt.plot(x, a)
plt.plot(x, b)
plt.show()
plt.savefig('final2270.jpg')
```
